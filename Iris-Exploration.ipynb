```python
sns.heatmap(df.isnull())
plt.show()
```

### Inference
1. There are no null values in the dataset

```python
sns.catplot(data=df,x='Species',kind='count').set(title='Graph showing the count for\neach category of Irises')
plt.show()
```

### Inference
1. All the species have equal number of rows

```python
sns.scatterplot(data=df,x='SepalLengthCm',y='SepalWidthCm',hue='Species')
plt.show()
```

## We can infer from the top that Iris-setosa has greater sepal width but smaller Sepal lentgh.Iris Versicolor has average sepal width and lentgh and Iris Virginica has has higher sepal lentgh but average sepal width

```python
sns.scatterplot(data=df,x='PetalLengthCm',y='PetalWidthCm',hue='Species')
plt.show()
```

## We can infer that Iris Setosa has very small petal width and Petal lentgh and in quite contrsat Iris Virginica has the highest Petal width and Petal length among the three but like before Iris Versicolor is still average in both attributes.

## The above inference can be better understood by the side by side bar plots 

```python
fig,axes=plt.subplots(2,2,figsize=(9,9))
sns.barplot(x=df['Species'],y=df['SepalLengthCm'],ax=axes[0,0])
sns.barplot(x=df.Species,y=df.SepalWidthCm,ax=axes[0,1])
sns.barplot(x=df['Species'],y=df['PetalLengthCm'],ax=axes[1,0])
sns.barplot(x=df['Species'],y=df['PetalWidthCm'],ax=axes[1,1])
fig.suptitle('Bar graph showing the relation between \nmean values of all the attributes and species',fontsize=30)
fig.show()
```

### Inference
1. As infered from this plots Iris Verginica has largest mean Sepal Length and mean Petal length among the three, It also has largest mean Petal width but has medium size Sepl width.
2. Iris Setosa has smallest mean petal length, mean sepal length and mean petal width but has largest mean sepal width.
3. Iris Versicolor has medium medium Sepal length,medium Petal length and medium Petal width but has medium smallest Sepal width

```python
fig,axes=plt.subplots(2,2,figsize=(9,9))
sns.boxplot(x=df['Species'],y=df['SepalLengthCm'],ax=axes[0,0])
sns.boxplot(x=df['Species'],y=df['SepalWidthCm'],ax=axes[0,1])
sns.boxplot(x=df['Species'],y=df['PetalLengthCm'],ax=axes[1,0])
sns.boxplot(x=df['Species'],y=df['PetalWidthCm'],ax=axes[1,1])
fig.suptitle("Box graph showing outliers for \nall the attributes based on species",fontsize=30)
plt.show()
```

### Inference:
1. Iris-Setosa has the smallest featues and is less distributed with some outliers
2. Iris-Versicolor has average features with not outliers
3. Iris-Verginica has largest featuers and is more distributed than setosa with some outliers
